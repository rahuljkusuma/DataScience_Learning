## Data_Ananlysis

```python
df = pd.read_csv('data_p7.csv')

view_count = df.view_count
likes = df.likes
ratio = df.ratio
```

```python
# df.head()
```

```python
plt.scatter(view_count, likes, s=100, edgecolors='black', linewidths=1, alpha=0.75)

plt.title('Trending Youtube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()
```

```python
plt.scatter(view_count, likes, s=100, edgecolors='black', linewidths=1, alpha=0.75)

plt.xscale('log')
plt.yscale('log')

plt.title('Trending Youtube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()
```

***Read documentation for "cmap".***

```python
plt.scatter(view_count, likes, c=ratio, cmap='summer', s=100,
            edgecolors='black', linewidths=1, alpha=0.75)

cbar = plt.colorbar()
cbar.set_label('Like/Dislike Ratio')

plt.xscale('log')
plt.yscale('log')

plt.title('Trending Youtube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()
```
