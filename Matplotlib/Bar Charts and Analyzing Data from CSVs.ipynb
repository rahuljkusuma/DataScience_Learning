## Data Analysis

```python
from collections import Counter
import csv
```

***Reading Data with csv module***

```python
with open('data_p2.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    
    language_counter = Counter()
    
    for row in csv_reader:
        language_counter.update(row['LanguagesWorkedWith'].split(';'))
```

```python
language_counter
```

```python
language_counter.most_common(15)
```

```python
languages = [i[0] for i in language_counter.most_common(15)]
popularity = [i[1] for i in language_counter.most_common(15)]
```

```python
languages
```

```python
popularity
```

```python
plt.bar(languages, popularity)

plt.title("Most Popular Languages")
plt.xlabel("Programming Languages")
plt.ylabel("Number Of Users")

plt.legend()

plt.tight_layout()
```

```python
plt.barh(languages, popularity) #For horizontal bar charts use "barh"

plt.title("Most Popular Languages")
# plt.ylabel("Programming Languages")
plt.xlabel("Number Of Users")

plt.legend()

plt.tight_layout()
```

```python
plt.barh(languages[::-1], popularity[::-1]) #For horizontal bar charts use "barh"

plt.title("Most Popular Languages")
# plt.ylabel("Programming Languages")
plt.xlabel("Number Of Users")

plt.legend()

plt.tight_layout()
```

***Reading Data with pandas module***

```python
import pandas as pd
```

```python
data = pd.read_csv('data_p2.csv')
```

```python
ids = data['Responder_id']
```

```python
lang_response = data['LanguagesWorkedWith']
```

```python
language_counter = Counter()

for response in lang_response:
    language_counter.update(response.split(';'))
```

```python
language_counter.most_common(15)
languages = [i[0] for i in language_counter.most_common(15)]
popularity = [i[1] for i in language_counter.most_common(15)]
```

```python
plt.barh(languages[::-1], popularity[::-1], color='#008fd5') #For horizontal bar charts use "barh"

plt.title("Most Popular Languages")
# plt.ylabel("Programming Languages")
plt.xlabel("Number Of Users")

plt.legend()

plt.tight_layout()
```
